```python
from groq import Groq
client = Groq()
```

```python
SYSTEM_PROMPT = """
You are a medical information assistant.
You provide:
- Medicine usage
- Dosage guidelines
- Side effects
- Safety warnings

Rules:
- Do NOT diagnose diseases
- Do NOT prescribe medication
- Always add a disclaimer
- Encourage consulting a doctor

End every response with:
"‚ö†Ô∏è This information is for educational purposes only. Consult a healthcare professional."
"""
```

```python
def medicine_chatbot(user_question):
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",  # ‚úÖ STABLE MODEL
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": user_question}
        ],
        temperature=0.3,
        max_tokens=500
    )
    
    return response.choices[0].message.content
```

```python
print(medicine_chatbot("What is Paracetamol used for?"))
```

```python
print(medicine_chatbot("What is Paracetamol used for?"))
```

```python
print(medicine_chatbot("Can I take paracetamol with ibuprofen?"))
print(medicine_chatbot("What are side effects of amoxicillin?"))
```

```python
while True:
    user_input = input("You: ")
    
    if user_input.lower() in ["exit", "quit"]:
        print("Bot: Take care üëã")
        break
        
    reply = medicine_chatbot(user_input)
    print("\nBot:", reply, "\n")
```
